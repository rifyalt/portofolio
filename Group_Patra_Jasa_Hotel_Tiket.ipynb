```python
def get_url(search_term):
    """Generate a url from a search term"""     # Tanggalnya bisa disesuaikan 2023-09-28 & checkout=2023-09-29
    template = "https://www.tiket.com/hotel/indonesia/{}?adult=2&room=1&checkin=2023-09-30&checkout=2023-`0-01&night=1" # disesuaikan dengan url
    search_term = search_term.replace(" ", "+")
    return template.format(search_term)
```
