```python
# Clean text
text_t = re.sub('[^A-Za-z0-9°]+', ' ', konten)
text_t = text_t.replace('\n', '').lower()
text_t
```

```python
def word_count(str):
    counts = dict()
    konten = str.split()

    for word in konten:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts

#print(word_count(text_c))
word_count(text_t)
```

```python
t = word_count(text_t)
```

```python
konten_ = pd.DataFrame(t.items(), columns=['Text', 'Count'])
#tujuan
```

```python
sum_konten = konten_.sort_values(by='Count', ascending=False)
sum_konten.head(10)
```
