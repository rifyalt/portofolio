# Availibilty & Rate Natal Paeriod

```python
def get_url(search_term):
    """Generate a url from a search term"""     # Tanggalnya bisa disesuaikan 2023-12-22 & checkout=2023-12-26
    template = "https://www.tiket.com/hotel/indonesia/{}?adult=2&room=1&checkin=2023-12-22&checkout=2023-12-26&night=4" # disesuaikan dengan url
    search_term = search_term.replace(" ", "+")
    return template.format(search_term)
```
